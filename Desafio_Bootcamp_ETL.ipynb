### Um Pipeline de ETL com Python
A idéia desse projeto é criar um pipeline simples de ETL. Nesse caso, a ideia aqui é, a partir de uma tabela com dados climáticos do país inteiro (fictícia, por enquanto), filtrar e disponibilizar dados da região nordeste para estudos posteriores.




# Etapa de Extração de dados

Nessa sessão, faremos o carregamento da biblioteca Pandas e do dataframe contendo as informações climáticas de todas as partes do país que iremos utilizar.

```python
# Carregando as bibliotecas
import pandas as pd

# Carregando o arquivo contendo os dados
df = pd.read_csv("dados_climaticos_brasil_ficticios.csv")

df.head()
```

# Etapa de Transformação dos dados

Como a intenção ao final do pipeline é fornecer informações do nordeste para análises posteriores, aqui faremos todos o processo de filtragem e seleção de features importantes.


```python
# Prinmeiro vamos filtrar somente os dados correspondentes a região nordeste do nosso dataframe
df_nordeste = df[df['regiao'] == "Nordeste"]

# Agora selecionaremos as coluna de interesse
df_nordeste = df_nordeste[['data','cidade','estado','temperatura_max_c','temperatura_min_c','precipitacao_mm','umidade_relativa_percent','condicao_climatica']]
df_nordeste.head()
```

# Etapa de Load(carregamento) dos dados

Após filtrar as informações de interesse, aqui iremos salvar esses dados em um novo aequivo .csv para disponibiliza-lo. Facilitando análises que necessitem informações climáticas somente da região nordeste.

```python
df_nordeste.to_csv("dados_climaticos_nordeste.csv", sep = ',', encoding = 'utf-8')
```
